```python
import math
import random
import numpy as np
def CLG(bits, a, b, full = False):
    assert bits % 2 == 0 or full
    assert len(a) == bits and len(b) == bits
    c = '0'
    cout = ''
    for i in range(len(a)):
        if a[bits - i - 1] == '1' and b[bits - i - 1] == '1':
            c = '1'
        elif a[bits - i - 1] == '0' and b[bits - i - 1] == '0':
            c = '0'
        
        cout = c + cout
    
    cout = cout + '0'
    if not full:
        cout = cout[0:int(len(a)/2) + 1]
        
    return cout

def xor_m(a, b, l):
    assert len(a) == l
    assert len(b) == l
    result = ''
    for i in range(l):
        if a[l-i-1] != b[l-i-1]:
            result = '1' + result
        else:
            result = '0' + result
    return result
```

```python
def approx_add(a, b, l, CLG_length):
    assert CLG_length % 2 == 0
    result = ''
    num1 = str(bin(a))[2:]
    num2 = str(bin(b))[2:]
    while(len(num1) < l):
        num1 = '0' + num1
    while(len(num2) < l):
        num2 = '0' + num2
    remain =  int(l % (CLG_length / 2))

    for i in range(math.floor(l / CLG_length * 2) - 1):
        if i == 0:
            s = xor_m(num1[(l-CLG_length):l], num2[(l-CLG_length):l], CLG_length)
            c = CLG(CLG_length, num1[(l-CLG_length):l], num2[(l-CLG_length):l], True)
            result = xor_m(s, c[1:], CLG_length) + result
        else:
            s = xor_m(num1[l-int(CLG_length/2)*(i+2):l-int(CLG_length/2)*(i+1)], num2[l-int(CLG_length/2)*(i+2):l-int(CLG_length/2)*(i+1)], int(CLG_length/2))
            c = CLG(CLG_length, num1[l-int(CLG_length/2)*(i+2):l-int(CLG_length/2)*(i)], num2[l-int(CLG_length/2)*(i+2):l-int(CLG_length/2)*(i)])
            result = xor_m(s[0:int(CLG_length/2)], c[1:], int(CLG_length/2)) + result
        
    if remain != 0:
        s = '0' + xor_m(num1[0:remain], num2[0:remain], remain)
        c = CLG(remain + int(CLG_length/2), num1[0:remain + int(CLG_length/2)], num2[0:remain + int(CLG_length/2)], True)
        result = xor_m(s[0:remain + 1], c[0:remain + 1], remain + 1) + result
    else:
        c = CLG(CLG_length, num1[0:CLG_length], num2[0:CLG_length])
        s = xor_m(num1[0:int(CLG_length/2)], num2[0:int(CLG_length/2)], int(CLG_length/2))
        if c[0] == '1':
            result = '1' + result
    return result
        
```

```python
def approx_add2(a, b, l, CLG_length):
    assert CLG_length % 2 == 0
    result = ''
    num1 = str(bin(a))[2:]
    num2 = str(bin(b))[2:]
    while(len(num1) < l):
        num1 = '0' + num1
    while(len(num2) < l):
        num2 = '0' + num2
    remain =  int(l % (CLG_length / 2))
    if remain == 0:
        remain = int(CLG_length/2)
    for i in range(math.ceil(l / CLG_length * 2) - 3):
        if i == 0:
            s = xor_m(num1[(l-CLG_length):l], num2[(l-CLG_length):l], CLG_length)
            c = CLG(CLG_length, num1[(l-CLG_length):l], num2[(l-CLG_length):l], True)
            result = xor_m(s, c[1:], CLG_length) + result
        else:
            s = xor_m(num1[l-int(CLG_length/2)*(i+2):l-int(CLG_length/2)*(i+1)], num2[l-int(CLG_length/2)*(i+2):l-int(CLG_length/2)*(i+1)], int(CLG_length/2))
            c = CLG(CLG_length, num1[l-int(CLG_length/2)*(i+2):l-int(CLG_length/2)*(i)], num2[l-int(CLG_length/2)*(i+2):l-int(CLG_length/2)*(i)])
            result = xor_m(s[0:int(CLG_length/2)], c[1:], int(CLG_length/2)) + result
    
    

    if l >= (CLG_length * 3 / 2):
        s = '0' + xor_m(num1[0:remain + int(CLG_length/2)], num2[0:remain + int(CLG_length/2)], remain + int(CLG_length/2))
        c = CLG(remain + int(CLG_length), num1[0:remain + int(CLG_length)], num2[0:remain + int(CLG_length)], True)
        result = xor_m(s[0:remain + int(CLG_length/2) + 1], c[0:remain + int(CLG_length/2) + 1], remain + int(CLG_length/2) + 1) + result
    else:
        s = '0' + xor_m(num1[0:remain + int(CLG_length)], num2[0:remain + int(CLG_length)], remain + int(CLG_length))
        c = CLG(remain + CLG_length, num1[0:remain + CLG_length], num2[0:remain + CLG_length], True)
        result = xor_m(s[0:remain + int(CLG_length) + 1], c[0:remain + int(CLG_length) + 1], remain + int(CLG_length) + 1) + result
    return result
```

```python
# final version of approximate adder
def approx_add3(a, b, l, block_size):
    result = ''
    num1 = str(bin(a))[2:]
    num2 = str(bin(b))[2:]
    while(len(num1) < l):
        num1 = '0' + num1
    while(len(num2) < l):
        num2 = '0' + num2
    block_num = math.ceil(l/block_size)
    c = '0'
    for i in range(block_num):
        if block_num == 1:
            temp_sum, c = sum_generate(num1[0:l], num2[0:l], c, block_size)
            result = c + temp_sum + result
        elif i == block_num - 1 and i != 0:
            if block_num <= 2:
                _, c = sum_generate(num1[l-block_size*i:l-block_size*(i-1)], num2[l-block_size*i:l-block_size*(i-1)], '0', block_size)
            else:
                _, c = sum_generate(num1[l-block_size*i:l-block_size*(i-2)], num2[l-block_size*i:l-block_size*(i-2)], '0', block_size*2)
            temp_sum, c = sum_generate(num1[0:l-block_size*i], num2[0:l-block_size*i], c, l-block_size*i)
            result = c + temp_sum + result
        else:
            temp_sum, _ = sum_generate(num1[l-block_size*(i+1):l-block_size*i], num2[l-block_size*(i+1):l-block_size*i], c, block_size)
            _, c = sum_generate(num1[l-block_size*(i+1):l-block_size*i], num2[l-block_size*(i+1):l-block_size*i], '0', block_size) 
            result = temp_sum + result
    return result
```

```python
def sum_generate(a,b,c,l):
    assert len(a) == l
    assert len(b) == l
    r = ""
    for i in range(l):
        if a[l-i-1] == '1' and b[l-i-1] == '1':
            r = c + r
            c = '1'
        elif a[l-i-1] == '0' and b[l-i-1] == '0':
            r = c + r
            c = '0'
        elif c == '1':
            r = '0' + r
            c = '1'
        else:
            r = '1' + r
            c = '0'
    return r, c
```

```python
# proposed adder with higher bits CLA and low bits OR adder
def low_bit_or(a, b, num_l, block_l, or_l):
    rest = num_l - or_l
    num1 = str(bin(a))[2:]
    num2 = str(bin(b))[2:]
    while(len(num1) < num_l):
        num1 = '0' + num1
    while(len(num2) < num_l):
        num2 = '0' + num2
    num11 = num1[:rest]
    num12 = num1[-or_l:]
    num21 = num2[:rest]
    num22 = num2[-or_l:]
    a_m = int(num11,2)
    b_m = int(num21,2)
    out1 = approx_add3(a_m,b_m,rest,block_l)
    out2 = ""
    for i in range(or_l):
        if num12[i] == "0" and num22[i] == "0":
            out2 = out2 + "0"
        else:
            out2 = out2 + "1"
    return int(out1+out2,2)
```

```python

num = 100000

for MAA in np.arange(0.9,1.1,0.01):
    count = 0
    for i in range(num):
        a = random.randint(0,2**32-1)
        b = random.randint(0,2**32-1)
        out= low_bit_or(a,b,32,4,0)
        error = abs(a + b - out)
        acc = 1 - error/(a + b)
        if acc >= MAA:
            count += 1
    print(MAA)
    print(count/num)
```
